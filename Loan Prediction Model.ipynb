```python
genders = ['Male', 'Female']
married = ['Yes','No']
education = ['Not Graduate', 'Graduate']
self_employed = ['Yes','No']
credit_history = ['1.0', '0.0']
property_area = ['Rural','Semiurban', 'Urban']
```

```python
enc = preprocessing.OneHotEncoder(categories=[genders, married, education,
                                              self_employed, credit_history, property_area], sparse=False)
```

```python
enc
```

```python
one_hot_encoded = enc.fit_transform(filtered_data.loc[:,['Gender','Married','Education', 'Self_Employed', 
                                       'Credit_History', 'Property_Area']].values)
```
